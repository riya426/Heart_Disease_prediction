```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
#from sklearn import svm
#from sklearn.tree import DecisionTreeClassifier
#from sklearn.ensemble import RandomForestClassifier
#from sklearn.neighbors import KNeighborsClassifier  # KNN classifier





from sklearn.metrics import accuracy_score
```

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
#model = DecisionTreeClassifier(random_state=2)
#model = RandomForestClassifier(n_estimators=100, random_state=2)
#model = KNeighborsClassifier(n_neighbors=5)  # You can experiment with the value of n_neighbors

  
```

```python
model.fit(X_train,Y_train)
```

```python

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on Training data : ", training_data_accuracy)
```

```python
import joblib
joblib.dump(model,'heart_disease_model.pkl')
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate confusion matrix
cm = confusion_matrix(Y_test, X_test_prediction)

# Check if the confusion matrix is correctly computed (optional)
print("Confusion Matrix:")
print(cm)

# Visualize the confusion matrix using a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False, 
            xticklabels=["Negative", "Positive"], yticklabels=["Negative", "Positive"])

# Labels and title for the plot
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix for Heart Disease Prediction")

# Show the plot
plt.show()
```
